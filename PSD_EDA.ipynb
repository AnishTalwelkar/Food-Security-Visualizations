## Calculating the stock to use ratio -
* Shows interrealtionship between the supply and demand. The lower the ratio, tighter the demand
* Formula --> Stock to use ratio = (Total Production - Total Use)/Total Use  * 100

```python
psd_filtered1['Stock to Use Ratio'] = (psd_filtered1['Ending Stocks']/psd_filtered1['Domestic Consumption'])*100
psd_filtered1
```

```python
sns.lineplot(x='End of Market Year',y='Stock to Use Ratio',hue='Commodity_Description',data=psd_filtered1).set(ylim=(0, 50))

##Outliers in corn ? Check 1960-75
```
